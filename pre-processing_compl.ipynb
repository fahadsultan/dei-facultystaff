```python
from glob import glob
import pandas as pd

fnames = glob('completions_data/2002-2007/*.csv')
all_df = []
year = 2002
for fname in fnames: 
    df = pd.read_csv(fname)
    df.columns = df.columns.str.upper()
    df.columns = [x.strip() for x in df.columns]
    df  = df[df['AWLEVEL'] == 5] #AW level 5 for bachelors degree
    df  = df[df['CRACE24']>0]
    df0 = df[(df['CIPCODE'] == 99)]
    df1 = df[(df['CIPCODE'] >=11.0101) & (df1['CIPCODE'] < 12)]
    

    df0['Year'] = year
    df1['Year'] = year
    df0 = df0[['UNITID','MAJORNUM', 'CIPCODE', 'CRACE24','CRACE18','CRACE21','CRACE16','Year']]
    df1 = df1[['UNITID','MAJORNUM', 'CIPCODE', 'CRACE24','CRACE18','CRACE21','CRACE16','Year']]
    year+=1
    all_df.append(df0)
    all_df.append(df1)
bigdf0 = pd.concat(all_df)
bigdf0 = bigdf0.rename(columns={'UNITID':'Unit Id'})
bigdf0 = bigdf0.rename(columns={'CIPCODE':'Degrees'})
bigdf0 = bigdf0.rename(columns={'CRACE24':'Grand total students'})
bigdf0 = bigdf0.rename(columns={'CRACE18':'Black total students'})
bigdf0 = bigdf0.rename(columns={'CRACE21':'Hispanic total students'})
bigdf0 = bigdf0.rename(columns={'CRACE16':'Grand total women students'})

```
