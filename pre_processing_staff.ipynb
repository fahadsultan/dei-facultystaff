```python
from glob import glob
import pandas as pd
fnames = glob('older_staff_data/*.csv')
all_df = []
year = 2002
for fname in fnames: 
    df = pd.read_csv(fname)
    df.columns = df.columns.str.lower()
    df.columns = [x.strip() for x in df.columns]
    df = df[(df['line'] == 18)]

    df['Year'] = year
    year+=1
    all_df.append(df)
bigdf = pd.concat(all_df)
bigdf.isna().sum()

bigdf.to_csv('older_staff.csv')

```

```python
from glob import glob
import pandas as pd
fnames = glob('old_staff_data/*.csv')
all_df = []
year = 2008
for fname in fnames: 
    df = pd.read_csv(fname)
    df.columns = df.columns.str.upper()
    df.columns = [x.strip() for x in df.columns]
    df = df[(df['SABDTYPE'] == 11)]
    df['Year'] = year
    year+=1
    all_df.append(df)



bigdf = pd.concat(all_df)
bigdf.to_csv("old_staff.csv")
```
