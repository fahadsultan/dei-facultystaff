```python
from glob import glob
import pandas as pd

fnames = glob('completions_data/2002-2007/*.csv')
all_df = []
year = 2002
for fname in fnames: 
    df = pd.read_csv(fname)
    df.columns = df.columns.str.upper()
    df.columns = [x.strip() for x in df.columns]
    df  = df[df['AWLEVEL'] == 9] #AW level 9 for doctor's degree
    df  = df[df['CRACE24']>0]
    df0 = df[(df['CIPCODE'] == 99)]
    df1 = df[(df['CIPCODE'] >=11.0101) & (df['CIPCODE'] < 12)]
    

    df0['Year'] = year
    df1['Year'] = year
    df0 = df0[['UNITID','MAJORNUM', 'CIPCODE', 'CRACE24','CRACE18','CRACE21','CRACE16','Year','AWLEVEL']]
    df1 = df1[['UNITID','MAJORNUM', 'CIPCODE', 'CRACE24','CRACE18','CRACE21','CRACE16','Year','AWLEVEL']]
    year+=1
    all_df.append(df0)
    all_df.append(df1)
bigdf0 = pd.concat(all_df)
bigdf0 = bigdf0.rename(columns={'UNITID':'Unit Id'})
bigdf0 = bigdf0.rename(columns={'CIPCODE':'Degrees'})
bigdf0 = bigdf0.rename(columns={'CRACE24':'Grand total students'})
bigdf0 = bigdf0.rename(columns={'CRACE18':'Black total students'})
bigdf0 = bigdf0.rename(columns={'CRACE21':'Hispanic total students'})
bigdf0 = bigdf0.rename(columns={'CRACE16':'Grand total women students'})

```

```python
from glob import glob
import pandas as pd
fnames = glob('completions_data/2011-2022/*.csv')
all_df = []
year = 2011
for fname in fnames: 
    df  = pd.read_csv(fname)
    df.columns = df.columns.str.upper()
    df.columns = [x.strip() for x in df.columns]
    df  = df[df['AWLEVEL'] > 16] #AW level 17 or higher for doctor's degree
    df  = df[df['CTOTALT']>0]
    df0 = df[(df['CIPCODE'] == 99)]
    df1 = df[(df['CIPCODE'] >=11.0101)]
    df1 = df1[(df1['CIPCODE'] < 12)]

    df0['Year'] = year
    df1['Year'] = year
    df0 = df0[['UNITID','MAJORNUM', 'CIPCODE', 'CTOTALT','CBKAAT','CHISPT','CTOTALW','Year','AWLEVEL']]
    df1 = df1[['UNITID','MAJORNUM', 'CIPCODE', 'CTOTALT','CBKAAT','CHISPT','CTOTALW','Year','AWLEVEL']]
    year+=1
    all_df.append(df0)
    all_df.append(df1)
bigdf2 = pd.concat(all_df)
bigdf2 = bigdf2.rename(columns={'UNITID':'Unit Id'})
bigdf2 = bigdf2.rename(columns={'CIPCODE':'Degrees'})
bigdf2 = bigdf2.rename(columns={'CTOTALT':'Grand total students'})
bigdf2 = bigdf2.rename(columns={'CBKAAT':'Black total students'})
bigdf2 = bigdf2.rename(columns={'CHISPT':'Hispanic total students'})
bigdf2 = bigdf2.rename(columns={'CTOTALW':'Grand total women students'})
```
