```python
from glob import glob
import pandas as pd

fnames = glob('enroll_data/1998, 2000, 2001/*.csv')
all_df = []
for fname in fnames: 
    df = pd.read_csv(fname)
    df.columns = df.columns.str.upper()
    df.columns = [x.strip() for x in df.columns]
    df  = df[df['LINE'] == 8] #Full time undergraduates

    year       = (fname[35:-4])
    df['Year'] = year
    df = df[['UNITID','EFRACE15', 'EFRACE03', 'EFRACE04','EFRACE09','EFRACE10','EFRACE16','Year']]
    df['Grand total enrolled']    = df['EFRACE15'] + df['EFRACE16']
    df['Black total enrolled']    = df['EFRACE03'] + df['EFRACE04']
    df['Hispanic total enrolled'] = df['EFRACE09'] + df['EFRACE10']
    df = df.rename(columns={'EFRACE16':'Total women enrolled'})
    df = df[['UNITID','Grand total enrolled', 'Black total enrolled', 'Hispanic total enrolled','Total women enrolled','Year']]
    all_df.append(df)
bigdf0 = pd.concat(all_df)
bigdf0
```
